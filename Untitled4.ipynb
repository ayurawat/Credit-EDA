```python
# Outliers are visible. Hence we are making a function to detect outliers and remove it thereafter

def remove_outlier(col):
    q25 = col.quantile(0.25) #lower quartile
    q75 = col.quantile(0.75) #upper quartile
    iqr = q75 - q25 #inter quartile range
    cutoff = iqr*1.5
    lower = q25 - cutoff
    upper = q75 + cutoff
    return lower, upper
#Remove outliers for AMT_INCOME_TOTAL
lower_0, upper_0 = remove_outlier(DF0.AMT_INCOME_TOTAL)
lower_1, upper_1 = remove_outlier(DF1.AMT_INCOME_TOTAL)

DF0= DF0.loc[(DF0.AMT_INCOME_TOTAL > lower_0) & (DF0.AMT_INCOME_TOTAL < upper_0)]
DF1=DF1.loc[(DF1.AMT_INCOME_TOTAL > lower_1) & (DF1.AMT_INCOME_TOTAL < upper_1)]
plt.figure(figsize=(18,10)) 

#plotting to view the result again
plt.subplot(1,2,1)
ax = sns.boxplot(y=DF0['AMT_INCOME_TOTAL'])
plt.title('Total income of customers without payment difficulties')

plt.subplot(1,2,2)
ax = sns.boxplot(y=DF1['AMT_INCOME_TOTAL'])
plt.title('Total income of customers with payment difficulties')
plt.show()
```

We can see that median value of clients in target group 0 ( i.e non defaulters) lies lower than 
target group 1, indicating that 2nd group had a higher median income of clients
